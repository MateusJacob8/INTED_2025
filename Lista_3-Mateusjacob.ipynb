```python
matrix = []

for _ in range(3):
    print("-------------------")
    row = []
    for _ in range(3):
        row.append(int(input("Número: ")))
    matrix.append(row)

is_identity = True
for l in range(3):
    for c in range(3):
        if l == c and matrix[l][c] != 1:
            is_identity = False
        elif l != c and matrix[l][c] != 0:
            is_identity = False

print("É identidade" if is_identity else "Não é identidade")
```

```python
matrix = []

for _ in range(3):
  print("-------------------")
  row = []
  for _ in range(3):
    row.append(int(input("Número: ")))
  matrix.append(row)

is_symetric = True
for l in range(3):
  for c in range(3):
    if not is_symetric: break
    if matrix[l][c] != matrix[c][l]: is_symetric = False
  if not is_symetric: break

print("É simétrica" if is_symetric else "Não simétrica")
```

```python
matrix1 = []
matrix2 = []

for _ in range(3):
  print("-------------------")
  row = []
  for _ in range(3):
    row.append(int(input("Número: ")))
  matrix1.append(row)

for _ in range(3):
  print("-------------------")
  row = []
  for _ in range(3):
    row.append(int(input("Número: ")))
  matrix2.append(row)

matrix_sum = []
matrix_sub = []

for l in range(3):
  sum_row = []
  sub_row = []
  for c in range(3):
    num1 = matrix1[l][c]
    num2 = matrix2[l][c]

    matrix_sum.append(num1 + num2)
    matrix_sub.append(num1 - num2)
  matrix_sum.append(sum_row)
  matrix_sub.append(sub_row)

print(matrix_sum)
print("----------")
print(matrix_sub)
```

```python
products = {
    "Macarrão": [25.90, 28.50, 24.99],
    "Maçã": [8.90, 9.50, 7.99],
    "banana": [45.90, 42.50, 50.00]
}
max_variation = 0
max_variation_product = ""

for p, prices in products.items():
  variation = max(prices) - min(prices)
  if variation > max_variation:
    max_variation = variation
    max_variation_product = p

print(f"Produto: {max_variation_product}. Variação: R$ {max_variation}")
```
